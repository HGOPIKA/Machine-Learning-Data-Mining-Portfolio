```python
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules

# Load the dataset (replace 'dataset.csv' with the path to your dataset)
df = pd.read_csv('Online Retail.csv')

# Data Preprocessing
# Assuming the dataset is already cleaned and ready for analysis
# Convert data into one-hot encoded format
basket = (df.groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))

# Convert quantity values to binary
basket = basket.applymap(lambda x: 1 if x > 0 else 0)

# Perform FP-growth
frequent_itemsets = fpgrowth(basket, min_support=0.05, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Display frequent itemsets and association rules
print("\nFrequent Itemsets:")
print(frequent_itemsets)

print("\nAssociation Rules:")
print(rules)
```

```python
import matplotlib.pyplot as plt

# Plotting frequent itemsets
plt.figure(figsize=(10, 6))
frequent_itemsets.plot(kind='bar', x='itemsets', y='support', color='blue')
plt.xlabel('Frequent Itemsets')
plt.ylabel('Support')
plt.title('Support of Frequent Itemsets')
plt.xticks(rotation=45)
plt.show()
```

```python
#Use find_frequent_patterns to find patterns that occur over the support threshold

patterns = fpg.find_frequent_patterns(labelcount['labels'], 2)
patternsdf = pd.DataFrame({'Label Association': list(patterns.keys()), 'Occurrences': list(patterns.values())})
```

```python
patternsdf
```

```python
rules = fpg.generate_association_rules(patterns, 0.10)
rulesdf = pd.DataFrame({'Association Rules': list(rules.keys()), 
                        'Labels': [0]*len(list(rules.keys())), 'Probabilities': list(rules.values())})
rulesdf.loc[:, 'Labels'] = rulesdf['Probabilities'].apply(lambda x: x[0][0])
rulesdf.loc[:, 'Probabilities'] = rulesdf['Probabilities'].apply(lambda x: x[1])
rulesdf = rulesdf.sort_values('Probabilities', ascending = False)
```

```python
from apyori import apriori # association rule
```

```python
records = []
for i in range(0, 315):
    records.append([str(df.values[i,j]) for j in range(0, 7)])
```

```python
# association rule apriori
association_rules = apriori(records, min_support=0.0045, min_confidence=0.7, min_lift=3, min_length=2)
association_results = list(association_rules)
```

```python
# print the total number of rules mined by the apriori class
print(len(association_results))
```

```python
# print the first item in the association rules
print(association_results[0])
```

```python
# print the rules, the support, the confidence, and lift for each rule
for item in association_results:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================") 
```
